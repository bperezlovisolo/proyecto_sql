```python
#cambiamos el nombre de la columna año para tener mayor claridad
df_invierno.rename(columns= {'AGNO': 'año'}, inplace=True)
df_invierno.columns
```

```python
#cambiamos el formato de los nombres de las columnas para hacer mas sencillo la referencia. 
df_invierno.columns = df_invierno.columns.str.lower().str.replace(" ", "_")
df_invierno.columns
```

```python
pd.set_option('display.max_columns', None)
```

```python
#funcion para eliminar columnas que no usaremos
def drop_columns (df, columns):
    df.drop(columns, axis=1, inplace=True)
```

```python
columns_drop = ["rbd", "dgv_rbd", "cod_pro_rbd", "nom_reg_rbd_a", "cod_pro_rbd", "cod_com_rbd", "cod_deprov_rbd", "cod_depe", "cod_depe2", "cod_ense", "cod_ense2", "let_cur", "mrun", "cod_com_alu"]

drop_columns(df_invierno, columns_drop)
```
