### FILTRAMOS COLUMNAS

```python
#funcion para eliminar columnas que no usaremos
def drop_columns (df, columns):
    df.drop(columns, axis=1, inplace=True)
```

```python
columns_drop = ["rbd", "dgv_rbd", "cod_pro_rbd", "nom_reg_rbd_a", "cod_pro_rbd", "cod_com_rbd", "cod_deprov_rbd", "cod_depe", "cod_depe2", "cod_ense", "cod_ense2", "let_cur", "mrun", "cod_com_alu"]

drop_columns(df_invierno, columns_drop)
```

```python
columns_drop = ["cod_reg_rbd", "cod_grado", "fec_nac_alu", "nom_com_alu"]

drop_columns(df_invierno, columns_drop)
```

```python
columns_drop = ["asis_promedio"]

drop_columns(df_invierno, columns_drop)
```

### FILTRAMOS POR PROVINCIAS

```python
df_invierno["nom_deprov_rbd"].value_counts()
```

```python
def clean_column(df, column_name):
    df_clean = df.copy()
    df_clean = df_clean[df_clean[column_name].isin(["ANTOFAGASTA - TOCOPILLA", "MAGALLANES"])]
    print(f"Valores únicos en '{column_name}':", 
          df_clean[column_name].unique())
    return df_clean
```

```python
df_prov = clean_column(df_invierno, "nom_deprov_rbd")
```

```python
df_prov.head()
```
