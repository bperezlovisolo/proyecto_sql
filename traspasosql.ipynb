```python
import mysql.connector
import pandas as pd
```

```python
import getpass  # To get the password without showing the input
password = getpass.getpass()
```

```python
conn = mysql.connector.connect(host='localhost',
                        user='root',
                        passwd=password)
```

```python
cursor = conn.cursor()

cursor
```

```python
cursor.execute("CREATE DATABASE IF NOT EXISTS ED_CHILE")
```

```python
cursor.execute("USE ED_CHILE")
```

```python
cursor.execute("""
    CREATE TABLE IF NOT EXISTS localidad (
        rbd INT,
        nom_rbd VARCHAR(255) NOT NULL,
        nom_com_rbd VARCHAR(255) NOT NULL,
        lon DECIMAL NOT NULL,
        lat DECIMAL NOT NULL,
        zona VARCHAR(255) NOT NULL
    )
""")
print("Table localidad created successfully!")
```

```python
df_sql= pd.read_csv("localizacion.csv")
```

```python
tupla=list(df_sql.itertuples(index=False,name=None))
```

```python
print(len(tupla[0]))
```

```python

cursor.executemany("INSERT INTO localidad (rbd , nom_rbd , nom_com_rbd, lon , lat, zona) VALUES (%s, %s, %s, %s, %s, %s)", tupla)
conn.commit()
print(f"{cursor.rowcount} items inserted!")
```
