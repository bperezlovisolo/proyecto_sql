Usaremos la función informe para generar un analisis exploratorio de la base de datos.

```python
def informe(x):
    print("Información del DataFrame:")
    print(x.info())
    print("Descripción estadística:")
    print(x.describe())
    print("Valores nulos:")
    print(x.isnull().sum())
    print("Tipos de datos:")
    print(x.dtypes)
    
    print("Primeras 5 filas:")
    print(x.head())
    print("Últimas 5 filas:")
    print(x.tail())
    print("Forma del DataFrame:")
    print(x.shape)
    print("columnas")
    print(x.columns)
    print("numericos")
    print(x.select_dtypes(include=["number"]).columns)
```

Investigando en https://www.bcn.cl/siit/nuestropais/
La República de Chile se sitúa en el margen suroccidental de América del Sur, entre los 17º 30' y 56º 30' de latitud sur.

y se podrian agrupar de la siguiente manera:

-17° a -30° S: Norte de Chile (Arica, Iquique, Antofagasta).
-33° S: Zona central (Santiago/Valparaíso).
-41° a -56° S: Zona sur y austral (Puerto Montt hasta Cabo de Hornos).

Longitud: Chile continental se mueve mayoritariamente entre los -70° y -75° de longitud oeste.

```python
def clasificar_zona(lat):
    if lat >= -26: return 'Norte Grande'
    elif lat >= -32: return 'Norte Chico'
    elif lat >= -36: return 'Zona Central'
    elif lat >= -41: return 'Zona Sur'
    else: return 'Zona Austral'

df_final['zona'] = df_final['lat'].apply(clasificar_zona)
```

Función para limpiar y dejar solo las provincias de interés Magallanes y Antofagasta

```python
def clean_column(df, column_name):
    df_clean = df.copy()
    df_clean = df_clean[df_clean[column_name].isin(["ANTOFAGASTA - TOCOPILLA", "MAGALLANES"])]
    print(f"Valores únicos en '{column_name}':", 
          df_clean[column_name].unique())
    return df_clean
```

```python
df_prov = clean_column(df_invierno, "nom_deprov_rbd")
```

### Unamos los datos a través de Merge

haremos la función a nuestros nuevos archivos de datos.

```python
df_locaciones= pd.read_csv("localizacion.csv") 
df_asistencia= pd.read_csv("asistencia.csv")
```

```python
def union(df1,df2):
    df_union=pd.merge(df1,df2,on=["nom_rbd","nom_com_rbd"],how='left')
    return df_union
```

```python
df_union = union(df_asistencia,df_locaciones)   
df_union.head()
```
